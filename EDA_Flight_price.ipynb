```python
df["Date"]= df['Date_of_Journey'].str.split("/").str[0]
df["Month"]= df['Date_of_Journey'].str.split("/").str[1]
df["Year"]= df['Date_of_Journey'].str.split("/").str[2]
df
```

```python
df['Date']= df['Date'].astype(int)
df['Month']= df['Month'].astype(int)
df['Year']= df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey', axis = 1 , inplace = True)
```

```python
df['Arrival_Time'].unique()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df["Arrival_Hr"]= df['Arrival_Time'].str.split(":").str[0]
df["Arrival_Mins"]= df['Arrival_Time'].str.split(":").str[1]
```

```python
df["Arrival_Hr"] = df["Arrival_Hr"].astype(int)
df["Arrival_Mins"] = df["Arrival_Mins"].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
df
```

```python
df["Dep_hr"]=df['Dep_Time'].str.split(":").str[0]
df["Dep_min"]=df['Dep_Time'].str.split(":").str[1]
```

```python
df["Dep_hr"]=df["Dep_hr"].astype(int)
df["Dep_min"]=df["Dep_min"].astype(int)
```

```python
df.drop('Dep_Time', axis= 1, inplace= True)
```

```python
df["Total_Stops"] = df["Total_Stops"].map({'non-stop':0,'1 stop':1, '2 stops':2, '3 stops':3,'4 stops':4})
```

```python
df["Total_Stops"] = df["Total_Stops"].astype(int)
```

```python
df.info()
```

```python
df.drop("Route", axis = 1, inplace = True)
```

```python
df['Duration_hrs'] = df['Duration'].str.split(" ").str[0].str.split('h').str[0]
df
```

```python
mask = df['Duration_hrs'].str.contains('m', na=False)
df.loc[mask, 'Duration_mins'] = df.loc[mask, 'Duration_hrs'].str.replace('m', '', regex=True)
df.loc[mask, 'Duration_hrs'] = 0
df["Duration_hrs"].unique()
```

```python
df['Duration_mins'] = df['Duration'].str.split(" ").str[1].str.split('h').str[0]
df
```

```python
df['Duration_mins']=df['Duration_mins'].str[0]
```

```python
df['Duration_mins']=df['Duration_mins'].fillna(0)
df
```

```python
df["Duration_hrs"]=df["Duration_hrs"].astype(int)
df["Duration_mins"]=df["Duration_mins"].astype(int)
```
